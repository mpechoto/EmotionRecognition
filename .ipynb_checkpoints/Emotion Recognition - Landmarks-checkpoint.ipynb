```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import os, random
```

### Loading database through pandas

This corpus contains images of the face of an actress expressing four different emotional states, beyond the neutral state: "joy", "anger", "fear", "sadness". The file "imagedb_CH_disciplina.csv" associates each image with an emotional label and also with the coordinates os characteristic points of the face.

```python
file = pd.read_csv("Faces/imagedb_CH_disciplina.csv", header=None)
```

```python
file.head()
```

```python
file = file.add_prefix("feat_")
file.rename(index=str,columns={"feat_0":"Image","feat_1":"Emotion"},inplace=True)
file.head()
```

Lets visualize the distribution of "Emotions".

As we can see all classes are balanced with 39 samples each one.

```python
file.Emotion.value_counts()
```

```python
from collections import Counter
pd.DataFrame.from_dict(Counter(file.Emotion),orient='index').plot(kind='bar')
```

Lets select only some classes: "neutral", "joy", "anger", "fear" and "sadness"

```python
dataset = file[file["Emotion"].isin(["neutral", "joy", "anger", "fear","sadness"])]
```

```python
dataset.Emotion.unique()
```

```python
dataset.head()
```

Now we have 195 samples divided in 5 classes and 55 features (columns from feat_9 to feat_64)

```python
dataset.shape
```

```python
dataset = dataset.drop(dataset.loc[:,'feat_2':'feat_8'],axis=1)
dataset.head()
```

```python
#replacements = {'('}
#dataset=dataset.apply(lambda x: x.replace(r"[\\(\\)]",""))
dataset=dataset.apply(lambda x: x.str.replace(r"[\\(\\)]",""))
dataset.head()
```

```python
dataset.head()
```

Split the coordinates in X and Y

```python
for i in range(9,65):
    dataset['feat_{}_x'.format(i)] = dataset.loc[:,'feat_{}'.format(i)].apply(lambda x: x.split(";")[0])    
    dataset['feat_{}_y'.format(i)] = dataset.loc[:,'feat_{}'.format(i)].apply(lambda x: x.split(";")[1])
    dataset = dataset.drop('feat_{}'.format(i),axis=1)
```

```python
dataset.head()
```

Lets see some examples from the database

```python
def plot_image(x,y,axis,image):
    img = mpimg.imread("Faces\\"+image)
    axis.imshow(img)
    axis.scatter(x,y,s=10,c='r')
#X = dataset.ix[0,'feat_9_x'::2]
#Y = dataset.ix[0,'feat_9_y'::2]
```

```python
files = os.listdir("Faces\\")

images = dataset.Image.tolist()
set(files) - set(images)
```

```python
match =  (set(files) & set(images) )
```

```python
while (random.choice(os.listdir("Faces\\")) not in list(match)):
       image_name = random.choice(os.listdir("Faces\\"))
```

```python
fig = plt.figure(figsize=(20,15))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
for i in range(9):
    image_name = random.choice(os.listdir("Faces\\"))
    while (image_name not in list(match)):
        image_name = random.choice(os.listdir("Faces\\"))
    new = dataset[dataset["Image"]==image_name]
    X = new.loc[:,'feat_9_x'::2]
    Y = new.loc[:,'feat_9_y'::2]
    ax = fig.add_subplot(3,3,i+1,xticks=[],yticks=[])
    plot_image(X,Y,ax,image_name)
```

Lets train!

```python
from sklearn.svm import SVC
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
```

```python
#features = dataset.loc[:,'feat_9_x':]
#labels = dataset.Emotion
train = dataset.loc[:,'feat_9_x':]
labels = dataset.Emotion
```

```python
from sklearn.model_selection import train_test_split

train_feat,test_feat,train_labels,test_labels = train_test_split(
    train,labels, test_size=0.25,random_state=42 )

train_feat.shape,test_feat.shape,train_labels.shape,test_labels.shape
#training = dataset[:int(len(dataset)*0.8)] #get first 80% of file list
#prediction = dataset[-int(len(dataset)*0.2):] #get last 20% of file list
```

```python
#Set the classifier as a support vector machines with polynomial kernel
clf = SVC(kernel='linear', C=1.0, probability=True, tol=1e-3,verbose=True)

clf.fit(train_feat,train_labels)
```

```python
y_pred = clf.predict(test_feat)
```

```python
y_pred
```

```python
clf.score(test_feat,test_labels)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test_labels,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score (clf, train,labels,cv=5)
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn import metrics

predicted = cross_val_predict(clf, train,labels, cv=5)
metrics.accuracy_score(labels, predicted) 
```

```python
C=1.0
models = (SVC(kernel='linear', C=C),
          svm.LinearSVC(C=C),
          SVC(kernel='rbf', gamma=0.7, C=C),
          SVC(kernel='poly', degree=3, C=C),
          KNeighborsClassifier(3),
          DecisionTreeClassifier(max_depth=5),
          RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),
          MLPClassifier(hidden_layer_sizes=(5,2),activation='tanh', alpha=1e-05,verbose=False,max_iter=1000),
          AdaBoostClassifier(),
          GaussianNB(),
         )

names = ["SVC with linear kernel","LinearSVC (linear kernel)","SVC with RBF kernel","SVC with polynomial (degree 3) kernel",
         "Nearest Neighbors","Decision Tree", "Random Forest", "Neural Net", "AdaBoost","Naive Bayes"]
         
#models = (clf.fit(train_feat,train_labels) for clf in models)
```
